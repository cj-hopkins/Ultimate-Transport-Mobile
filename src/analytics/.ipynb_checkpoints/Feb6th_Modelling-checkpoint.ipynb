### Create train and test dataframes and build a Random Forest LM

```python
#Split the dataset into two datasets: 70% training and 30% test.
shuffled = shuffle(df_rf)

y = shuffled[['time_bin']]
X_train, X_test, y_train, y_test = train_test_split(shuffled, y, test_size=0.3)

#Extract only the columns we are using as predictors 
X= X_train[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]
X_train=X_train[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]
X_test=X_test[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]

y_train = y_train['time_bin']     # extract the target feature 
y_test= y_test['time_bin']
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42) # Instantiate model with 1000 decision trees
# rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(X_train, y_train)    # Train the model on training data
```

### Run the model on the training and test and print the metrics

```python
def print_metrics(predictions,actual):   # Evaluation function
    errors = abs(predictions - actual)   # absolute errors
    mae= np.mean(abs(errors))
    ape = abs(errors / actual) * 100      # MAPE
    accuracy = 100 - np.mean(ape)
    print('MSE:', round(np.mean(errors**2), 2))  #MSE
    print('MAE:',mae)
    print('MAPE:',round(np.mean(ape),2), '%')
    print('Accuracy:', round(accuracy, 2), '%')
```

```python
print ('\nRANDOM FOREST')
print ('Training set:')
print_metrics(rf.predict(X_train),y_train )
print ('\nTest set:')
print_metrics(rf.predict(X_test),y_test )
```

### Gradient Boosting Model

```python
#Split the dataset into two datasets: 70% training and 30% test.
shuffled = shuffle(df_rf)

y = shuffled[['time_bin']]
X_train, X_test, y_train, y_test = train_test_split(shuffled, y.values.ravel(), test_size=0.3)

#Extract only the columns we are using as predictors 
X= X_train[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]
X_train=X_train[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]
X_test=X_test[['Precipitation_None','Precipitation_Slight','Precipitation_Moderate','time_interval_8am-10am','time_interval_5pm-7pm','direction','dwell_time']]
```

```python
est = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0, loss='ls').fit(X_train, y_train)
est
```

```python
print ('\nGRADIENT BOOSTING')
print ('Training set:')
print_metrics(est.predict(X_train),y_train)
print ('\nTest set:')
print_metrics(est.predict(X_test),y_test)
```

```python
params = {'n_estimators': 100, 'max_depth': 1,
          'learning_rate': 0.1, 'loss': 'ls'}

# compute test set deviance
test_score = np.zeros((params['n_estimators'],), dtype=np.float64)

for i, y_pred in enumerate(est.staged_predict(X_test)):
    test_score[i] = est.loss_(y_test, y_pred)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, est.train_score_, 'b-',
         label='Training Set Deviance')
plt.plot(np.arange(params['n_estimators']) + 1, test_score, 'r-',
         label='Test Set Deviance')
plt.legend(loc='upper right')
plt.xlabel('Boosting Iterations')
plt.ylabel('Deviance')

# #############################################################################
# Plot feature importance
feature_importance = est.feature_importances_
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
plt.subplot(1, 2, 2)
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, df_rf.columns)
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```
